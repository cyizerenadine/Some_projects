```python
def info(house_url):
    response = requests.get(house_url)
    soup = BeautifulSoup(response.text, 'lxml')
    
    title = soup.find("div", class_ = "perfetto-page-title property-banner has-image-bg").find("div", class_ = "container").select("span")[0].text
    prop_type = soup.select("span.property-type")[0].text
    advert = soup.select("span.status-label")[0].text.strip().split('/')[1].strip()
    price = soup.select("span.property-price-number")[0].text + soup.select("span.price-prefix")[1].text
    temp = soup.select("span.property-meta-number")[0].text
    if temp == "N/A" or int(temp) <= 15:
        plot = "N/A"
        bedrooms = temp 
        bathrooms =temp
    else:
        plot = soup.select("span.property-meta-number")[0].text + ' ' + soup.select("span.property-meta-name")[0].text
        bedrooms = soup.select("span.property-meta-number")[1].text 
        bathrooms =soup.select("span.property-meta-number")[2].text 
   
    address = soup.select("div.col-md-12 ")[0].text.strip('\n')
    
    data_dict = {"Title":title, "Advert type": advert, "Price": price, "Bedrooms": bedrooms, "Address": address, "Property type": prop_type,
                "Bathrooms": bathrooms, "Plot size": plot}
    
    return data_dict
```

```python
def final_data(urls):
    data = []
    for link in urls:
        data.append(info(link))
    df = pd.DataFrame(data)
    df.to_csv("plut_properties.csv")
    return df
```
